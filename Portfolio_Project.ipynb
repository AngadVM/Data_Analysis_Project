## Clean up Data

### Want to view only USA races,50km or 50mi, 2018

```python
df[df['Event distance/length'] == '50mi']
```

```python
#combine both 50km and 50mi
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event']== 2018) ]
```

```python
df[df['Event name']== 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

### combine all the filters together

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event']== 2018)  & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event']== 2018)  & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head()
```

```python
df2.shape
```

### remove (USA) from event name


```python
df2['Event name']= df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

### clean up athlete age

```python
df2['athlete_age'] = 2018- df2['Athlete year of birth']
```

### remove 'h' from athlete performance

```python
df2['Athlete performance'] = df2['Athlete performance'].str.replace('h', '')
```

```python
df2.head()
```

### drop columns
dropping columns which are not needed such as Athlete club, Athlete country, Athlete year of birth, Athlete age category

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head()
```

### clean up null values

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

### check for duplicates

```python
df2[df2.duplicated()== True]  # no duplicates found
```

### reset index

```python
df2.reset_index(drop=True)
```

### fix types

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2['Athlete ID'] = df2['Athlete ID'].astype(int)
```

```python
df2.dtypes
```

```python
df2.head()
```

### renaming columns

```python
df2 = df2.rename(columns = {
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_avg_speed',
    'Athlete ID': 'athlete_id'

})
```

```python
df2.head()
```

### reorder columns

```python
df3 = df2[['race_day','athlete_id','race_name','athlete_age','race_number_of_finishers','race_length','athlete_gender','athlete_performance','athlete_avg_speed','year']]
```

```python
df3.head()
```

### find 2 races that a person ran in 2018 - Sarasota/ Everglades


```python
df3[df3['race_name'] == 'Everglades 50 Mile Ultra Run ']
```

### people taking part in marathons at the age of 50 and above

```python
df3[df3['athlete_age'] >= 50]
```

## Difference in speed for the 50k, 50mi male to female

### What age groups are the best in 50km Race (show first 10 values)

```python
df3.query('race_length == "50km"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count > 19').head(10)
```

### What age groups are the worst in 50km Race (show first 10 values)

```python
df3.query('race_length == "50km"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending=True).query('count > 19').head(10)
```

### Seasons for the data -> SLower in summer than winter?

* spring 3-5
* summer 6-8
* fall 9-11
* winter 12-2

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```

```python
df3.head(25)
```
